## Training YOLOv8 on Vehicle Registration Plate

This project trains YOLOv8 on custom data which is the vehicle license plate. The trained model is then used for detecting number plate on moving vehicles.

Dataset source: https://www.kaggle.com/datasets/fareselmenshawii/large-license-plate-dataset?resource=download

```python
PATH = '/content/drive/MyDrive/mydata/ComputerVision/VehicleRegistrationPlate'
```

```python
import os
from ultralytics import YOLO
```

```python
# load model
model = YOLO("yolov8n.yaml")
```

```python
# Train model
results = model.train(data=os.path.join(PATH, "config.yaml"), epochs=50)
```

```python
# save model results

import shutil

source_path = '/content/runs'
destination_path = '/content/drive/MyDrive/mydata/ComputerVision/VehicleRegistrationPlate/run50'

# Copy the contents of the source directory to the destination directory
shutil.copytree(source_path, destination_path)
```
